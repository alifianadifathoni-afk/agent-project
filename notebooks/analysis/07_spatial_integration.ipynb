# Spatial Integration Project

Tracker for spatial data integration: field boundaries + soil + NDVI.

## Objectives

- Download field boundaries for analysis area
- Download SSURGO soil data (texture, drainage, organic matter, pH)
- Acquire satellite imagery and calculate NDVI
- Integrate all data layers for comprehensive field analysis
- Create interactive dashboard map

## Data Sources

- **Field Boundaries**: Michigan corn fields (USDA NASS Crop Sequence Boundaries style)
- **Soil Data**: USDA NRCS SSURGO (Soil Survey Geographic Database)
- **NDVI**: Sentinel-2 satellite imagery (Copernicus)

## Data Integration

```python
import pandas as pd
import geopandas as gpd

# Load merged spatial data
fields = gpd.read_file('data/michigan_spatial_integration.geojson')
print("Integrated Spatial Data:")
print(fields[['field_id', 'county', 'area_acres', 'soil_type', 'drainage_class', 'ndvi_mean']].to_string(index=False))
```

## Key Findings

### Soil Characteristics
- **MI2600001 & MI2600002**: Moderately well drained soils with good agricultural potential
- **MI2600003**: Poorly drained Saginaw clay soils - may require drainage management

### Vegetation Index (NDVI)
- Peak season (July 2024) NDVI values range from 0.65 to 0.80
- Highest NDVI in MI2600002 (Gratiot) - indicates optimal crop health
- Lower NDVI in MI2600003 correlates with poorly drained soils

### Data Integration Benefits
- Combining soil drainage class with NDVI helps identify underperforming areas
- Fields with high clay content (MI2600003) show lower vegetation vigor
- Interactive map enables field-level analysis for management decisions
